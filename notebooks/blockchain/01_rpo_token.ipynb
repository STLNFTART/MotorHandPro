## Token Burn Mechanism

```python
def token_burn_simulation(days=365, initial_supply=1e9, burn_rate=0.001):
    t = np.arange(days)
    supply = initial_supply * (1 - burn_rate) ** t
    return t, supply

t, supply = token_burn_simulation()

plt.figure(figsize=(12, 6))
plt.plot(t, supply / 1e9, 'b-', linewidth=2)
plt.xlabel('Days')
plt.ylabel('Token Supply (Billions)')
plt.title('RPO Token Supply with Burn Mechanism', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.show()

print(f'Initial supply: {supply[0]/1e9:.2f}B tokens')
print(f'Supply after 1 year: {supply[-1]/1e9:.2f}B tokens')
print(f'Reduction: {(1 - supply[-1]/supply[0])*100:.2f}%')
```
