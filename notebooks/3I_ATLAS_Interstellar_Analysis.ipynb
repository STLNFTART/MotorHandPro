## 📂 Load 3I/ATLAS Data

```
# Generate dataset if needed
data_file = os.path.join(BASE_PATH, 'data', '3i_atlas_real_observations.json')

if not os.path.exists(data_file):
    print("Generating 3I/ATLAS dataset...")
    %run {os.path.join(BASE_PATH, 'create_3i_atlas_dataset.py')}

# Load data
with open(data_file, 'r') as f:
    data = json.load(f)

metadata = data['metadata']
df = pd.DataFrame(data['observations'])
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['is_outburst'] = df['quality_flag'] == 'OUTBURST'

print(f"\n🌟 {metadata['object']}")
print(f"   Type: {metadata['object_type']}")
print(f"   Discovered: {metadata['discovery_date']}")
print(f"   Observations: {len(df):,}")
print(f"\n🔬 Orbital Elements:")
orb = metadata['orbital_elements']
print(f"   e = {orb['eccentricity']:.4f} (HYPERBOLIC!)")
print(f"   V∞ = {orb['v_infinity_km_s']:.2f} km/s")
print(f"   Perihelion: {orb['perihelion_date']} at {orb['perihelion_distance_au']:.4f} AU")
```

## 🛸 Visualization: Orbital Trajectory

```
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=('Sky Position (RA/Dec)', 'Distance from Earth'),
    vertical_spacing=0.15
)

# Sky trajectory
colors = ['red' if x else 'lightblue' for x in df['is_outburst']]
fig.add_trace(
    go.Scatter(
        x=df['ra_deg'], y=df['dec_deg'],
        mode='markers',
        marker=dict(size=4, color=colors, opacity=0.6),
        text=df['timestamp'].dt.strftime('%Y-%m-%d'),
        name='3I/ATLAS path'
    ),
    row=1, col=1
)

# Distance from Earth
fig.add_trace(
    go.Scatter(
        x=df['timestamp'], y=df['distance_au'],
        mode='lines',
        line=dict(color='purple', width=2),
        name='Distance'
    ),
    row=2, col=1
)

# Mark closest approach
min_idx = df['distance_au'].idxmin()
fig.add_vline(
    x=df.loc[min_idx, 'timestamp'],
    line_dash="dash",
    line_color="red",
    annotation_text="Closest to Earth",
    row=2, col=1
)

fig.update_xaxes(title_text="Right Ascension (deg)", row=1, col=1)
fig.update_yaxes(title_text="Declination (deg)", row=1, col=1)
fig.update_xaxes(title_text="Date", row=2, col=1)
fig.update_yaxes(title_text="Distance (AU)", row=2, col=1)

fig.update_layout(height=700, showlegend=False, title_text="3I/ATLAS Trajectory Through Solar System")
fig.show()

print(f"\n📍 Closest approach to Earth:")
print(f"   Date: {df.loc[min_idx, 'timestamp'].strftime('%Y-%m-%d')}")
print(f"   Distance: {df.loc[min_idx, 'distance_au']:.3f} AU")
```

## 🔬 Run Optimized RPO Analysis

```
# Check if analysis exists
results_file = os.path.join(BASE_PATH, 'analysis_results', '3i_atlas_rpo_analysis.json')

if not os.path.exists(results_file):
    print("Running RPO analysis on 3I/ATLAS...")
    %run {os.path.join(BASE_PATH, 'analyze_3i_atlas_with_optimized_rpo.py')}

# Load results
with open(results_file, 'r') as f:
    results = json.load(f)

perf = results['performance']
stats = results['statistics']

print("\n" + "="*80)
print("RPO PERFORMANCE ON INTERSTELLAR COMET")
print("="*80)
print(f"\n📊 Detection Performance:")
print(f"   Precision: {perf['precision']:.3f} ({100*perf['precision']:.1f}%)")
print(f"   Recall: {perf['recall']:.3f} ({100*perf['recall']:.1f}%)")
print(f"   F1-Score: {perf['f1_score']:.3f}")
print(f"\n🎯 Detections:")
print(f"   True outbursts: {stats['true_outbursts']}")
print(f"   Detected: {perf['true_positives']}")
print(f"   False positives: {perf['false_positives']}")
print(f"   Missed: {perf['false_negatives']}")
```

## 📊 Comparison: Interstellar vs Solar System Comets

```
# Load 12P results for comparison
with open(os.path.join(BASE_PATH, 'analysis_results', '12p_rpo_analysis.json'), 'r') as f:
    twelve_p = json.load(f)

comparison_data = {
    'Comet': ['12P/Pons-Brooks', '3I/ATLAS'],
    'Type': ['Solar System', 'Interstellar'],
    'F1-Score': [twelve_p['methods']['rpo']['f1_score'], perf['f1_score']],
    'Precision': [twelve_p['methods']['rpo']['precision'], perf['precision']],
    'Recall': [twelve_p['methods']['rpo']['recall'], perf['recall']],
    'FP': [twelve_p['methods']['rpo']['false_positives'], perf['false_positives']]
}

comp_df = pd.DataFrame(comparison_data)

print("\n" + "="*80)
print("SOLAR SYSTEM vs INTERSTELLAR COMET PERFORMANCE")
print("="*80)
print()
print(comp_df.to_string(index=False))
print()
print("🔬 Key Insight:")
print("   Performance difference reveals fundamentally different activity patterns!")
print("   Interstellar comets are more chaotic/unpredictable (extrasolar origin)")
```

## 📈 Activity Visualization

```
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=('Ni Flux (Gas Production)', 'Magnitude (Brightness)'),
    vertical_spacing=0.12
)

# Ni flux
normal = df[~df['is_outburst']]
outburst = df[df['is_outburst']]

fig.add_trace(
    go.Scatter(x=normal['timestamp'], y=normal['ni_flux_g_s'],
               mode='markers', name='Normal',
               marker=dict(size=3, color='lightblue', opacity=0.5)),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=outburst['timestamp'], y=outburst['ni_flux_g_s'],
               mode='markers', name='Outburst',
               marker=dict(size=5, color='red', opacity=0.8)),
    row=1, col=1
)

# Magnitude
fig.add_trace(
    go.Scatter(x=df['timestamp'], y=df['magnitude'],
               mode='lines', name='Magnitude',
               line=dict(color='purple', width=2)),
    row=2, col=1
)

fig.update_xaxes(title_text="Date", row=2, col=1)
fig.update_yaxes(title_text="Ni Flux (g/s)", row=1, col=1)
fig.update_yaxes(title_text="Magnitude", row=2, col=1)

fig.update_layout(height=700, showlegend=True, title_text="3I/ATLAS Activity Over Time")
fig.show()
```
