## Cardiac Model Simulation

```python
# Simulate cardiac model
def cardiac_model(t, heart_rate=70):
    # Simple cardiac cycle model
    period = 60 / heart_rate
    ecg = np.sin(2 * np.pi * t / period)
    ecg += 0.3 * np.sin(4 * np.pi * t / period)
    return ecg

t = np.linspace(0, 3, 1000)
ecg = cardiac_model(t)

plt.figure(figsize=(14, 5))
plt.plot(t, ecg, 'b-', linewidth=2)
plt.xlabel('Time (s)')
plt.ylabel('ECG Signal')
plt.title('Simulated Cardiac Cycle', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.show()
```
