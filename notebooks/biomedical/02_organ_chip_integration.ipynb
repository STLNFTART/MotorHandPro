## Organ-Chip Data Analysis

```python
def simulate_organ_chip_data(hours=24):
    t = np.linspace(0, hours, 1000)
    # Cell viability
    viability = 100 - 5 * (1 - np.exp(-t / 10)) + np.random.randn(1000) * 2
    # Metabolic rate
    metabolism = 50 + 20 * np.sin(2 * np.pi * t / 24) + np.random.randn(1000) * 3
    return t, viability, metabolism

t, viability, metabolism = simulate_organ_chip_data()

fig, axes = plt.subplots(2, 1, figsize=(14, 8))
axes[0].plot(t, viability, 'b-', linewidth=2)
axes[0].set_ylabel('Cell Viability (%)')
axes[0].set_title('Organ-Chip Cell Viability', fontweight='bold')
axes[0].grid(True, alpha=0.3)

axes[1].plot(t, metabolism, 'g-', linewidth=2)
axes[1].set_xlabel('Time (hours)')
axes[1].set_ylabel('Metabolic Rate')
axes[1].set_title('Metabolic Activity', fontweight='bold')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
