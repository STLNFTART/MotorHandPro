## Experiment Tracking

```python
class LabAssistant:
    def __init__(self):
        self.experiments = []
        
    def create_experiment(self, name, params):
        exp = {'name': name, 'params': params, 'results': None}
        self.experiments.append(exp)
        return exp
        
    def run_experiment(self, exp_id):
        exp = self.experiments[exp_id]
        # Simulate experiment
        exp['results'] = np.random.randn(100)
        return exp['results']

assistant = LabAssistant()
exp = assistant.create_experiment('Primal Logic Test', {'KE': 0.5, 'lambda': 0.16905})
results = assistant.run_experiment(0)
plt.plot(results)
plt.title('Experiment Results')
plt.show()
```
