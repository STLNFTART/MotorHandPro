## Temporal Field Simulation

```python
def temporal_field(t, tau=5.0, lambda_val=0.16905):
    return np.exp(-lambda_val * (t / tau))

t = np.linspace(0, 50, 1000)
field = temporal_field(t)

plt.figure(figsize=(12, 5))
plt.plot(t, field, 'b-', linewidth=2)
plt.xlabel('Time (s)')
plt.ylabel('Field Strength')
plt.title('Temporal Displacement Field', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.show()
```
