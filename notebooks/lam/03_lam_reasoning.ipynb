## Action Planning Example

```python
class ActionPlanner:
    def __init__(self):
        self.goals = []
        self.actions = []
        
    def add_goal(self, goal):
        self.goals.append(goal)
        
    def plan(self):
        # Simple planning algorithm
        for goal in self.goals:
            action = {'goal': goal, 'status': 'planned'}
            self.actions.append(action)
        return self.actions

planner = ActionPlanner()
planner.add_goal('Book flight to Mars')
planner.add_goal('Reserve hotel')
actions = planner.plan()
for a in actions:
    print(f'Action: {a}')
```
