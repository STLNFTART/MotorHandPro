## Quantum-Inspired Optimization

```python
def quantum_optimization(f, x0, iterations=100, h=0.01):
    x = x0
    history = [x]
    
    for _ in range(iterations):
        # Quantum-inspired gradient
        grad = (f(x + h) - f(x - h)) / (2 * h)
        # Update with quantum tunneling probability
        x = x - 0.1 * grad + 0.05 * np.random.randn()
        history.append(x)
    
    return np.array(history)

# Optimize quadratic function
f = lambda x: (x - 3)**2
history = quantum_optimization(f, 0)

plt.figure(figsize=(12, 5))
plt.plot(history, 'b-', linewidth=2, alpha=0.7)
plt.axhline(y=3, color='r', linestyle='--', label='True minimum')
plt.xlabel('Iteration')
plt.ylabel('x')
plt.title('Quantum-Inspired Optimization', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
