## Parameter Sensitivity Heatmap

```python
import seaborn as sns

# Generate parameter sweep data
KE_vals = np.linspace(0.1, 2.0, 20)
lambda_vals = np.linspace(0.05, 0.5, 20)
KE_grid, lambda_grid = np.meshgrid(KE_vals, lambda_vals)

# Compute metric (e.g., settling time)
metric = 1.0 / (KE_grid * lambda_grid + 0.1)

plt.figure(figsize=(10, 8))
sns.heatmap(metric, xticklabels=np.round(KE_vals, 2)[::2], 
            yticklabels=np.round(lambda_vals, 2)[::2],
            cmap='RdYlGn_r', annot=False)
plt.xlabel('KE')
plt.ylabel('Lambda')
plt.title('Parameter Sensitivity Heatmap', fontweight='bold', fontsize=14)
plt.show()
```
