## Quantum State Space

```python
def quantum_state(x, y, t):
    # Simplified quantum state
    r = np.sqrt(x**2 + y**2)
    return np.exp(-r/10) * np.sin(r - t)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, t in enumerate([0, np.pi/2, np.pi]):
    Z = quantum_state(X, Y, t)
    im = axes[i].contourf(X, Y, Z, levels=20, cmap='RdBu')
    axes[i].set_title(f't = {t:.2f}', fontweight='bold')
    axes[i].set_xlabel('x')
    axes[i].set_ylabel('y')
    plt.colorbar(im, ax=axes[i])

plt.tight_layout()
plt.show()
```
