## Sensor Data Simulation

```python
# Simulate sensor data
def generate_sensor_data(duration=10, dt=0.01, noise=0.1):
    t = np.arange(0, duration, dt)
    # Position sensor
    position = np.sin(t) + noise * np.random.randn(len(t))
    # Velocity sensor
    velocity = np.cos(t) + noise * np.random.randn(len(t))
    return t, position, velocity

t, pos, vel = generate_sensor_data()

fig, axes = plt.subplots(2, 1, figsize=(14, 8))
axes[0].plot(t, pos, 'b-', alpha=0.7, linewidth=1)
axes[0].set_ylabel('Position')
axes[0].set_title('Position Sensor Data', fontweight='bold')
axes[0].grid(True, alpha=0.3)

axes[1].plot(t, vel, 'r-', alpha=0.7, linewidth=1)
axes[1].set_xlabel('Time (s)')
axes[1].set_ylabel('Velocity')
axes[1].set_title('Velocity Sensor Data', fontweight='bold')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
