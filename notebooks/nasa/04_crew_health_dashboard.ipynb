## Health Metrics Simulation

```python
def simulate_health_metrics(days=500):
    t = np.arange(days)
    # Various health metrics
    heart_rate = 70 + 5 * np.sin(t / 30) + np.random.randn(days) * 2
    bone_density = 100 - 0.01 * t + np.random.randn(days) * 0.5
    muscle_mass = 100 - 0.02 * t + np.random.randn(days) * 0.5
    return t, heart_rate, bone_density, muscle_mass

t, hr, bd, mm = simulate_health_metrics()

fig, axes = plt.subplots(3, 1, figsize=(14, 10))
axes[0].plot(t, hr, 'r-', alpha=0.7)
axes[0].set_ylabel('Heart Rate (bpm)')
axes[0].set_title('Heart Rate Monitor', fontweight='bold')
axes[0].grid(True, alpha=0.3)

axes[1].plot(t, bd, 'b-', alpha=0.7)
axes[1].set_ylabel('Bone Density (%)')
axes[1].set_title('Bone Density', fontweight='bold')
axes[1].grid(True, alpha=0.3)

axes[2].plot(t, mm, 'g-', alpha=0.7)
axes[2].set_xlabel('Mission Day')
axes[2].set_ylabel('Muscle Mass (%)')
axes[2].set_title('Muscle Mass', fontweight='bold')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Experiment 1: Comprehensive Analysis

### Objective
Assess impact of different preprocessing techniques

### Methodology
Subject-leave-one-out validation protocol

### Expected Outcomes
Identify optimal preprocessing pipeline

### Parameters
- Sample Rate: 1000 Hz
- Window Size: 200 ms

```python
# Experiment 1 Setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix

print(f"======================================================================")
print(f"EXPERIMENT 1")
print(f"======================================================================")
print()

# Initialize experiment parameters
np.random.seed(42 + 1)
experiment_id = f"EXP_001"
results = {}

```

```python
# Generate/Load experimental data
n_samples = 1000
n_features = 9
n_classes = 4

# Simulate EMG-like data
X = np.random.randn(n_samples, n_features) * 0.5
y = np.random.randint(0, n_classes, n_samples)

# Add realistic patterns
for i in range(n_classes):
    mask = y == i
    X[mask] += np.random.randn(n_features) * (i + 1) * 0.3

print(f"Data shape: {X.shape}")
print(f"Classes: {np.unique(y)}")
print(f"Class distribution: {np.bincount(y)}")

```

```python
# Process data
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Normalize
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Train set: {X_train.shape}")
print(f"Test set: {X_test.shape}")

```

```python
# Train RandomForest classifier
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train
model.fit(X_train_scaled, y_train)

# Evaluate
train_score = model.score(X_train_scaled, y_train)
test_score = model.score(X_test_scaled, y_test)

print(f"Model: RandomForest")
print(f"Train accuracy: {train_score:.4f}")
print(f"Test accuracy: {test_score:.4f}")

results['model'] = 'RandomForest'
results['train_acc'] = train_score
results['test_acc'] = test_score

```

```python
# Visualize results
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Confusion matrix
y_pred = model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_pred)

axes[0].imshow(cm, cmap='Blues')
axes[0].set_title(f'Confusion Matrix - Experiment 1')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('True')

# Add text annotations
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        axes[0].text(j, i, cm[i, j], ha='center', va='center')

# Feature importance or coefficients
if hasattr(model, 'feature_importances_'):
    importance = model.feature_importances_
    axes[1].bar(range(len(importance)), importance)
    axes[1].set_title('Feature Importance')
    axes[1].set_xlabel('Feature Index')
    axes[1].set_ylabel('Importance')
else:
    # Plot decision boundary projection (first 2 features)
    from sklearn.decomposition import PCA
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_test_scaled)

    scatter = axes[1].scatter(X_pca[:, 0], X_pca[:, 1], c=y_test, cmap='viridis', alpha=0.6)
    axes[1].set_title('PCA Projection (First 2 Components)')
    axes[1].set_xlabel('PC1')
    axes[1].set_ylabel('PC2')
    plt.colorbar(scatter, ax=axes[1])

plt.tight_layout()
plt.savefig(f'experiment_001_results.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\n✓ Experiment 1 complete")
print(f"Results: {results}")

```
