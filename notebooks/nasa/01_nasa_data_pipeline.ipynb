## 1. NASA Data Sources

```python
# Simulate NASA data
def generate_spe_data(days=500):
    t = np.arange(days)
    # Random SPE events
    spe_events = np.zeros(days)
    event_times = np.random.choice(days, size=5, replace=False)
    for et in event_times:
        spe_events[et:min(et+3, days)] = np.random.exponential(50, min(3, days-et))
    return t, spe_events

t, spe = generate_spe_data()

plt.figure(figsize=(14, 5))
plt.plot(t, spe, 'r-', linewidth=2)
plt.xlabel('Mission Day')
plt.ylabel('SPE Dose Rate (mSv/day)')
plt.title('Solar Particle Events During Mission', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.show()
```
