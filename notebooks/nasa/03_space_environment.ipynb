## Radiation Environment

```python
def van_allen_radiation(altitude_km):
    # Simplified Van Allen belt model
    if 1000 < altitude_km < 6000:
        return 100 * np.exp(-(altitude_km - 3000)**2 / 1000**2)
    return 1

altitudes = np.linspace(0, 10000, 1000)
radiation = [van_allen_radiation(a) for a in altitudes]

plt.figure(figsize=(12, 5))
plt.plot(altitudes, radiation, 'r-', linewidth=2)
plt.xlabel('Altitude (km)')
plt.ylabel('Radiation Level (mSv/day)')
plt.title('Van Allen Radiation Belts', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.show()
```
