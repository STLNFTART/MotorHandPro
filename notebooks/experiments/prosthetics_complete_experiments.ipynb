## Experiment 1: Complete Dataset Analysis

Comprehensive analysis of all 5 EMG datasets with statistical validation.


```python
# Load and analyze all EMG datasets
from lam.integrations.dataset_loader import EMGDatasetLoader
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

loader = EMGDatasetLoader()
datasets = loader.list_datasets()

print("EMG Dataset Analysis")
print("=" * 70)

results = []
for ds in datasets:
    print(f"\nAnalyzing: {ds['name']}")
    try:
        data_dict = loader.load_dataset(ds['id'])
        stats = loader.get_dataset_statistics(ds['id'])
        
        results.append({
            'Dataset': ds['name'],
            'Subjects': stats['subjects'],
            'Gestures': stats['gestures'],
            'Channels': stats['num_channels'],
            'Samples': stats['total_samples'],
            'Mean': stats['data_range']['mean'],
            'Std': stats['data_range']['std']
        })
    except Exception as e:
        print(f"  Error: {e}")

# Create comparison dataframe
df_comparison = pd.DataFrame(results)
print("\nDataset Comparison:")
print(df_comparison.to_string(index=False))

# Visualize
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Subject count
axes[0,0].bar(range(len(df_comparison)), df_comparison['Subjects'])
axes[0,0].set_title('Number of Subjects per Dataset')
axes[0,0].set_xticks(range(len(df_comparison)))
axes[0,0].set_xticklabels([d[:20] for d in df_comparison['Dataset']], rotation=45, ha='right')

# Plot 2: Gesture count
axes[0,1].bar(range(len(df_comparison)), df_comparison['Gestures'], color='orange')
axes[0,1].set_title('Number of Gestures per Dataset')
axes[0,1].set_xticks(range(len(df_comparison)))
axes[0,1].set_xticklabels([d[:20] for d in df_comparison['Dataset']], rotation=45, ha='right')

# Plot 3: Channel count
axes[1,0].bar(range(len(df_comparison)), df_comparison['Channels'], color='green')
axes[1,0].set_title('Number of Channels per Dataset')
axes[1,0].set_xticks(range(len(df_comparison)))
axes[1,0].set_xticklabels([d[:20] for d in df_comparison['Dataset']], rotation=45, ha='right')
axes[1,0].set_yscale('log')

# Plot 4: Sample count
axes[1,1].bar(range(len(df_comparison)), df_comparison['Samples'], color='red')
axes[1,1].set_title('Total Samples per Dataset')
axes[1,1].set_xticks(range(len(df_comparison)))
axes[1,1].set_xticklabels([d[:20] for d in df_comparison['Dataset']], rotation=45, ha='right')
axes[1,1].set_yscale('log')

plt.tight_layout()
plt.savefig('dataset_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✓ Dataset analysis complete")
```
